Sentiment Analysis based on [this article](https://huggingface.co/mdhugol/indonesia-bert-sentiment-classification)

```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import pandas as pd

pretrained= "mdhugol/indonesia-bert-sentiment-classification"
pretrained2 = "ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa"

model = AutoModelForSequenceClassification.from_pretrained(pretrained)
tokenizer = AutoTokenizer.from_pretrained(pretrained)

model2 = AutoModelForSequenceClassification.from_pretrained(pretrained2)
tokenizer2 = AutoTokenizer.from_pretrained(pretrained2)
```

```python
sentiment_analysis = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
sentiment_analysis2 = pipeline("sentiment-analysis", model=model2, tokenizer=tokenizer2)

label_index = {'LABEL_0': 'positive', 'LABEL_1': 'neutral', 'LABEL_2': 'negative'}

#pos_text = "Sangat bahagia hari ini"
#neg_text = "Dasar anak sialan!! Kurang ajar!!"

#result = sentiment_analysis(pos_text)
#status = label_index[result[0]['label']]
#score = result[0]['score']
#print(f'Text: {pos_text} | Label : {status} ({score * 100:.3f}%)')

#result = sentiment_analysis(neg_text)
#status = label_index[result[0]['label']]
#score = result[0]['score']
#print(f'Text: {neg_text} | Label : {status} ({score * 100:.3f}%)')
```

```python
input_text = "ketua umum" ##kalimatnya masukin sini
```

```python
result = sentiment_analysis(input_text)
status = label_index[result[0]['label']]
score = result[0]['score']

result2 = sentiment_analysis2(input_text)
status2 = result2[0]['label']
score2 = result2[0]['score']

print(f"Text: {input_text}")
print(f'mdhugol model \t| Label : {status} ({score * 100:.3f}%)')
print(f'indobert model \t| Label : {status2} ({score2 * 100:.3f}%)')
```

```python
tweetdata = pd.read_csv(r"dataset/processed/processed_data.csv")
```

```python
for i in range(len(tweetdata)):

  input_text = tweetdata.at[i,"text_cleaned_stemmed"]

  result = sentiment_analysis(input_text)
  tweetdata.at[i,"mdhugol_sentiment"] = label_index[result[0]['label']]
  tweetdata.at[i,"mdhugol_sentiment_score"] = result[0]['score']

  result2 = sentiment_analysis2(input_text)
  tweetdata.at[i,"indobert_sentiment"] = result2[0]['label']
  tweetdata.at[i,"indobert_sentiment_score"] = result2[0]['score']
```

```python
tweetdata.to_csv(r"dataset/processed/processed_data_sentiment.csv")
```
